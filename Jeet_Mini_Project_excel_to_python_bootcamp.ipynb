# KPI Analysis (Q3)

Key Performance Indicators (KPIs) provide a quick overview of the business performance. In this section, we calculate total sales, total profit, total orders, average sales, average profit, and the maximum and minimum sales values.

```python
# Display all column names
print(df.columns.tolist())
```

```python
# KPI Calculations

df['Profit'] = df['TotalAmount'] - df['Tax'] - df['ShippingCost']

total_sales = df["TotalAmount"].sum()

total_profit = df["Profit"].sum()

total_orders = df["OrderID"].nunique()

average_sales = df["TotalAmount"].mean()

average_profit = df["Profit"].mean()

maximum_sales = df["TotalAmount"].max()

minimum_sales = df["TotalAmount"].min()
```

```python
kpi_df = pd.DataFrame({
    "KPI": [
        "Total Sales",
        "Total Profit",
        "Total Orders",
        "Average Sales",
        "Average Profit",
        "Maximum Sale",
        "Minimum Sale"
    ],
    "Value": [
        total_sales,
        total_profit,
        total_orders,
        average_sales,
        average_profit,
        maximum_sales,
        minimum_sales
    ]
})

kpi_df
```

```python
kpi_df["Value"] = kpi_df["Value"].apply(
    lambda x: f"{x:,.2f}" if isinstance(x, (int, float)) else x
)

kpi_df
```

### KPI Interpretation

The calculated KPIs provide an overview of the company's overall business performance. These indicators will serve as the baseline for further analysis of sales, customer behavior, product performance, and regional trends.

# Sales Analysis by State (Q4)

This analysis identifies the total sales generated by each state and helps determine the highest and lowest performing regions.

```python
# Total Sales by State

state_sales = (
    df.groupby("State")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

state_sales
```

```python
print("Highest Sales State:")
print(state_sales.idxmax())

print("\nHighest Sales Amount:")
print(state_sales.max())

print("\nLowest Sales State:")
print(state_sales.idxmin())

print("\nLowest Sales Amount:")
print(state_sales.min())
```

```python
plt.figure(figsize=(12,6))

state_sales.plot(kind="bar")

plt.title("Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

### Interpretation

The bar chart highlights the contribution of each state to the overall sales. States with consistently high sales may represent strong markets, while low-performing states may require targeted marketing campaigns, improved logistics, or promotional strategies.

# Category Analysis (Q5)

This section analyzes the total sales generated by each product category to identify the highest revenue-contributing categories.

```python
# ==========================================
# SALES BY CATEGORY
# ==========================================

category_sales = (
    df.groupby("Category")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

category_sales
```

```python
print("Highest Revenue Category:", category_sales.idxmax())
print("Sales:", category_sales.max())

print("\nLowest Revenue Category:", category_sales.idxmin())
print("Sales:", category_sales.min())
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    category_sales,
    labels=category_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Category-wise Sales")

plt.show()
```

### Interpretation

The pie chart illustrates the percentage contribution of each category to the total sales. Categories with larger shares contribute significantly to overall revenue and should remain a focus for inventory planning and marketing.

# Product Analysis (Q7 & Q18)

Since the dataset does not contain a "Sub-Category" column, Product Name analysis is performed to identify the highest-selling products.

```python
product_sales = (
    df.groupby("ProductName")["TotalAmount"]
      .sum()
      .sort_values(ascending=False)
)

product_sales.head(10)
```

```python
plt.figure(figsize=(12,6))

product_sales.head(10).plot(kind="barh")

plt.title("Top 10 Products by Sales")

plt.xlabel("Sales")

plt.ylabel("Product")

plt.tight_layout()

plt.show()
```

### Interpretation

The top-selling products contribute a significant proportion of total revenue. These products should receive priority in inventory management, marketing campaigns, and promotional strategies.

# Payment Method Analysis (Q11)

# Monthly Sales Trend (Q12)

This section evaluates monthly sales performance to identify seasonal trends and peak business periods.

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

df["Month"] = df["OrderDate"].dt.to_period("M")
```

```python
monthly_sales = (
    df.groupby("Month")["TotalAmount"]
      .sum()
)

monthly_sales
```

```python
plt.figure(figsize=(12,5))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

### Interpretation

The monthly sales trend helps identify peak sales periods and seasonal demand. These insights can assist management in planning inventory, staffing, and promotional campaigns more effectively.

# High Sales but Low Profit Products

Products generating high revenue but relatively low profit may indicate excessive discounts, high shipping costs, or narrow profit margins.

```python
product_analysis = (
    df.groupby("ProductName")
      .agg(
          TotalSales=("TotalAmount", "sum"),
          TotalProfit=("Profit", "sum")
      )
)

product_analysis.sort_values(
    by="TotalSales",
    ascending=False
).head(10)
```

```python
product_analysis.sort_values(
    by=["TotalSales", "TotalProfit"],
    ascending=[False, True]
).head(10)
```

### Possible Reasons

- Heavy discounts
- High logistics costs
- Expensive procurement
- Competitive pricing
- Low profit margin products
